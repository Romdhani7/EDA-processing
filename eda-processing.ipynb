# Bivariante Analysis
analyse  relationships between two continuous features.

```python
fig , axes=plt.subplots(nrows=2 , ncols=1 , figsize=(10,8))

sns.scatterplot(data=df, x="Confidence_Limit_Low", y="Confidence_Limit_High", color="crimson", ax=axes[0])

axes[0].set_xlabel(axes[0].get_xlabel(), fontsize=18)
axes[0].set_ylabel(axes[0].get_ylabel(), fontsize=18)
sns.scatterplot(data=df, x="Data_Value_Alt", y="Data_Value", color="crimson", ax=axes[1])

axes[1].set_xlabel(axes[1].get_xlabel(), fontsize=18)
axes[1].set_ylabel(axes[1].get_ylabel(), fontsize=18)
```

This  scatterplots '1,2) are  approximately a straight lines.
This means there is a linear relationship ( y = ax + b ) between these two features.
We can say these features are dependent because the value of one feature tells you the value of another

wa can use also The correlation matrix to visualize relationships between features.

```python
correlation_matrix = df[Num].corr()
plt.figure(figsize=(18,9))
sns.heatmap(correlation_matrix, cmap="coolwarm_r",annot_kws={'fontsize':15}, annot=True, vmin=-1, vmax=1)
plt.show()
```
